```python
import numpy as np
import matplotlib.pyplot as plt
from panqec.codes import Planar2DCode, Toric2DCode, RotatedPlanar2DCode, Color3DCode
from panqec.codes import Toric3DCode, Planar3DCode, RotatedPlanar3DCode, RotatedToric3DCode, Color488Code, Color666ToricCode

def average_stabilizers_per_qubit(code_class, lattice_size):
    #print(code_class)
    code = code_class(lattice_size)
    dense_Hx = code.Hx.toarray()
    stabilizer_counts = np.sum(dense_Hx, axis=1)
    avg_stabilizers = np.mean(stabilizer_counts)
    return avg_stabilizers

def plot_average_stabilizers(max_lattice_size):
    lattice_sizes = list(range(2, max_lattice_size + 1))
    code_classes = [
        (Planar2DCode, "Planar2DCode"), 
        (Toric2DCode, "Toric2DCode"), 
        (RotatedPlanar2DCode, "RotatedPlanar2DCode"), 
        #(Color3DCode, "Color3DCode"),
        (Toric3DCode, "Toric3DCode"), 
        (Planar3DCode, "Planar3DCode"), 
        (RotatedPlanar3DCode, "RotatedPlanar3DCode"), 
        #(RotatedToric3DCode, "RotatedToric3DCode"), 
        (Color488Code, "Color488Code"), 
        (Color666ToricCode, "Color666ToricCode")
    ]

    plt.figure(figsize=(12, 8))

    for code_class, label in code_classes:
        avg_stabilizers = [average_stabilizers_per_qubit(code_class, size) for size in lattice_sizes]
        plt.plot(lattice_sizes, avg_stabilizers, marker='o', label=label)

    plt.xlabel('Lattice Size')
    plt.ylabel('Average number of qubits per stabilizer')
    plt.title('Average number of qubits per stabilizer')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
plot_average_stabilizers(10)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from panqec.codes import Planar2DCode, Toric2DCode, RotatedPlanar2DCode, Color3DCode
from panqec.codes import Toric3DCode, Planar3DCode, RotatedPlanar3DCode, RotatedToric3DCode, Color488Code, Color666ToricCode

def average_stabilizers_per_qubit(code_class, lattice_size):
    #print(code_class)
    code = code_class(lattice_size)
    dense_Hx = code.Hx.toarray()
    stabilizer_counts = np.sum(dense_Hx, axis=0)
    avg_stabilizers = np.mean(stabilizer_counts)
    return avg_stabilizers

def plot_average_stabilizers(max_lattice_size):
    lattice_sizes = list(range(2, max_lattice_size + 1))
    code_classes = [
        (Planar2DCode, "Planar2DCode"), 
        (Toric2DCode, "Toric2DCode"), 
        (RotatedPlanar2DCode, "RotatedPlanar2DCode"), 
        #(Color3DCode, "Color3DCode"),
        (Toric3DCode, "Toric3DCode"), 
        (Planar3DCode, "Planar3DCode"), 
        (RotatedPlanar3DCode, "RotatedPlanar3DCode"), 
        #(RotatedToric3DCode, "RotatedToric3DCode"), 
        (Color488Code, "Color488Code"), 
        (Color666ToricCode, "Color666ToricCode")
    ]

    plt.figure(figsize=(12, 8))

    for code_class, label in code_classes:
        avg_stabilizers = [average_stabilizers_per_qubit(code_class, size) for size in lattice_sizes]
        plt.plot(lattice_sizes, avg_stabilizers, marker='o', label=label)

    plt.xlabel('Lattice Size')
    plt.ylabel('Average number of stabilizers per qubit')
    plt.title('Average number of stabilizers per qubit')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
plot_average_stabilizers(10)
```

```python
def num_qubits(code_class, lattice_size):
    code = code_class(lattice_size)
    return code.Hx.shape[0]

def plot_average_stabilizers(max_lattice_size):
    lattice_sizes = list(range(2, max_lattice_size + 1))
    code_classes = [
        (Planar2DCode, "Planar2DCode"), 
        (Toric2DCode, "Toric2DCode"), 
        (RotatedPlanar2DCode, "RotatedPlanar2DCode"), 
        #(Color3DCode, "Color3DCode"),
        (Toric3DCode, "Toric3DCode"), 
        (Planar3DCode, "Planar3DCode"), 
        (RotatedPlanar3DCode, "RotatedPlanar3DCode"), 
        #(RotatedToric3DCode, "RotatedToric3DCode"), 
        (Color488Code, "Color488Code"), 
        (Color666ToricCode, "Color666ToricCode")
    ]

    plt.figure(figsize=(12, 8))

    for code_class, label in code_classes:
        avg_stabilizers = [num_qubits(code_class, size) for size in lattice_sizes]
        plt.plot(lattice_sizes, avg_stabilizers, marker='o', label=label)

    plt.xlabel('Lattice Size')
    plt.ylabel('Number of Qubits')
    plt.title('Number of Qubits vs Lattice Size')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_average_stabilizers(10)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from panqec.codes import Planar2DCode, Toric2DCode, RotatedPlanar2DCode, Color3DCode
from panqec.codes import Toric3DCode, Planar3DCode, RotatedPlanar3DCode, RotatedToric3DCode, Color488Code, Color666ToricCode

def plot_sparsity_pattern(code_class, lattice_size):
    code = code_class(lattice_size)
    dense_Hx = code.Hx.toarray()

    plt.figure(figsize=(8, 6))
    plt.spy(dense_Hx, markersize=5)
    plt.title(f"Sparsity Pattern of {code_class.__name__} (Lattice Size: {lattice_size})")
    plt.xlabel("Qubits")
    plt.ylabel("Stabilizers")
    plt.grid(True)
    plt.show()

def plot_all_sparsity_patterns(lattice_size):
    code_classes = [
        (Planar2DCode, "Planar2DCode"), 
        (Toric2DCode, "Toric2DCode"), 
        (RotatedPlanar2DCode, "RotatedPlanar2DCode"), 
        #(Color3DCode, "Color3DCode"),
        (Toric3DCode, "Toric3DCode"), 
        (Planar3DCode, "Planar3DCode"), 
        (RotatedPlanar3DCode, "RotatedPlanar3DCode"), 
        #(RotatedToric3DCode, "RotatedToric3DCode"), 
        (Color488Code, "Color488Code"), 
        (Color666ToricCode, "Color666ToricCode")
    ]

    for code_class, label in code_classes:
        plot_sparsity_pattern(code_class, lattice_size)

# Example usage:
plot_all_sparsity_patterns(3)
```
